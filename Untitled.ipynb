```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customers.csv')
df.head()
```

```python
df.columns = ['CID','Gender','Age','AnnInc','SScore']
df.head()
```

```python
df.sample(20)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
x = df.iloc[:,-2:]
x.head()
```

```python
x.describe()
```

```python
sns.scatterplot(data = x,x='AnnInc',y='SScore')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
wcss = []
for i in range(1,21):
    model = KMeans(n_clusters=i)
    model.fit(x)
    wcss.append(model.inertia_)
print('WCSS')
print(wcss)

```

```python
plt.plot(list(range(1,21)),wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.xticks(list(range(1,21)))
plt.show()
```

```python
k5 = KMeans(n_clusters=5)
k5.fit(x)
```

```python
ypred_k5 = k5.predict(x)
print(len(ypred_k5))
```

```python
df['Cls5'] =ypred_k5
df['Cls5'].value_counts()
```

```python
cents = k5.cluster_centers_
print(cents)
```

```python
sns.scatterplot(data = df,x='AnnInc',y='SScore',hue= 'Cls5',palette='Set1')
plt.scatter(cents[:,0],cents[:,1],label='Centroids',marker = '*',s=120,color = 'black')
plt.legend()
plt.show()
```
